## ステージの結合

```python
import numpy as np
import pandas as pd
```

```python
train_df = pd.read_csv("../../data/Processed/train2.csv")
test_df = pd.read_csv("../../data/Processed/test2.csv")

stage_df = pd.read_csv("../../data/Raw/stage.csv")
print(stage_df.shape)
stage_df.head(10)
```

```python
print(train_df.shape)
train_df.head(10)
```

```python
train_df = pd.merge(train_df, stage_df, left_on="stage", right_on="stage")
test_df = pd.merge(test_df, stage_df, left_on="stage", right_on="stage")
```

```python
print(train_df.shape)
print(test_df.shape)
```

```python
train_df.sort_values('id', inplace=True)
test_df.sort_values('id', inplace=True)
```

```python
train_df.to_csv("../../data/Processed/train2.csv", index=False)
test_df.to_csv("../../data/Processed/test2.csv", index=False)
```
