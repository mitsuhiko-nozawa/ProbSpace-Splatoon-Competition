```python
import numpy as np
import pandas as pd
pd.set_option("display.max_columns",300)
```

```python
cols = ["category1", "category2", "key", "subweapon", "special", "mainweapon", "reskin", "[ja-JP]"]
buki_df = pd.read_csv("../../data/Raw/statink-weapon2.csv", usecols=cols)
main_df = pd.read_csv("../../data/Raw/main.csv")
sub_df = pd.read_csv("../../data/Raw/sub.csv")
special_df = pd.read_csv("../../data/Raw/special.csv")
```

```python
for name in buki_df["[ja-JP]"].unique():
    if name not in main_df["main"].unique():
        print(name)
```

```python
for name in main_df["special_"].unique():
    if name not in special_df["special_"].unique():
        print(name)
```

```python
for name in main_df["sub"].unique():
    if name not in sub_df["sub"].unique():
        print(name)
```

```python
buki_df
```

```python
buki_df = pd.merge(buki_df, main_df, left_on="[ja-JP]", right_on="main", how="outer", suffixes=["", "_"]).drop(columns="main")
buki_df = pd.merge(buki_df, sub_df, left_on="sub", right_on="sub", how="outer").drop(columns="sub")
buki_df = pd.merge(buki_df, special_df, left_on="special_", right_on="special_", how="outer").drop(columns="special_")

buki_df = buki_df.drop(columns=["[ja-JP]", "genre-main"])
```

```python
buki_df.to_csv("../../data/Processed/buki2.csv", index=False)
```
