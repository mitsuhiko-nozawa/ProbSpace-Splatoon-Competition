## 武器の外部データの結合
元データのweaponを主キーとして武器データをくっつける

武器が欠損している場合は全て欠損させた

```python
import numpy as np
import pandas as pd
```

```python
train_df = pd.read_csv("../../data/Processed/train1.csv")
test_df = pd.read_csv("../../data/Raw/test_data.csv")
buki_df = pd.read_csv("../../data/Raw/statink-weapon2.csv", usecols=["category1", "category2", "key", "subweapon", "special", "mainweapon", "reskin"])
print(buki_df.shape)
buki_df.head(10)
```

```python
print(train_df.shape)
train_df.head(10)
```

```python
def merge_buki(df, buki_df):
    res = pd.merge(df, buki_df, left_on="A1-weapon", right_on="key", how="outer", suffixes=["", ""]).drop(columns='key')
    res = pd.merge(res, buki_df, left_on="A2-weapon", right_on="key", how="outer", suffixes=["", "-A2"]).drop(columns='key')
    res = pd.merge(res, buki_df, left_on="A3-weapon", right_on="key", how="outer", suffixes=["", "-A3"]).drop(columns='key')
    res = pd.merge(res, buki_df, left_on="A4-weapon", right_on="key", how="outer", suffixes=["", "-A4"]).drop(columns='key')
    res = pd.merge(res, buki_df, left_on="B1-weapon", right_on="key", how="outer", suffixes=["", "-B1"]).drop(columns='key')
    res = pd.merge(res, buki_df, left_on="B2-weapon", right_on="key", how="outer", suffixes=["", "-B2"]).drop(columns='key')
    res = pd.merge(res, buki_df, left_on="B3-weapon", right_on="key", how="outer", suffixes=["", "-B3"]).drop(columns='key')
    res = pd.merge(res, buki_df, left_on="B4-weapon", right_on="key", how="outer", suffixes=["-A1", "-B4"]).drop(columns='key')
    return res
```

```python
train_df = merge_buki(train_df, buki_df)
test_df = merge_buki(test_df, buki_df)
```

```python
print(train_df.shape)
print(test_df.shape)
```

```python
train_df.sort_values('id', inplace=True)
test_df.sort_values('id', inplace=True)
```

```python
train_df.to_csv("../../data/Processed/train2.csv", index=False)
test_df.to_csv("../../data/Processed/test2.csv", index=False)
```
